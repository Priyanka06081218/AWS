```python
#Brute force
nums = [5, 3, 5, 6, 2, 1, 9, 4, 0]
n=len(nums)
k=3
rotations = k%n
for i in range(0, rotations):
  e=nums.pop()
  nums.insert(0,e)

print(nums)
```

```python
#Python slicing
nums = [5, 3, 5, 6, 2, 1, 9, 4, 0]
n=len(nums)
k=5
nums[:]=nums[n-k:]+nums[:n-k]
print(nums)
```

```python
#Without python slicing
nums = [5, 3, 5, 6, 2, 1, 9, 4, 0]
n=len(nums)
k=1
def reverse(nums, left, right):
  while left<right:
      nums[left], nums[right]=nums[right], nums[left]
      left+=1
      right-=1

reverse(nums, n-k, n-1)
reverse(nums, 0, n-k-1)
reverse(nums, 0, n-1)

print(nums)
```
